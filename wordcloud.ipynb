```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import wordcloud as wc
import matplotlib.pyplot as plt
```

```python
# df = pd.read_csv('dataset/wordcloud.csv',encoding='latin-1')
df = pd.read_csv('dataset/wordcloud-b4.csv',encoding='latin-1')
df.head()
```

```python
stopwords = set(line.strip() for line in open('stopwords.txt'))
stopwords = stopwords.union(set(['mr','br','mrs','one','two','said','comment','dtype','object','Name','hotel']))
wordcloud = WordCloud(
    width = 500,
    height = 400,
#     color_func=lambda *args, **kwargs: "black",
    colormap = 'Blues',
    background_color = 'white',
    min_font_size = 2,
    collocations=False, 
    stopwords = stopwords).generate(str(df['comments']))
fig = plt.figure(
    figsize = (50, 40),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.title('Hotel Tentrem')
plt.axis('off')
plt.tight_layout(pad=0)
plt.savefig("plot/WordCloud-Hotel-Novotel.png")
plt.show()
```
